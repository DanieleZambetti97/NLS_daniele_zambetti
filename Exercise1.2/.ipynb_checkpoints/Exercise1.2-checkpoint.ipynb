## Exercise 01.2
- Extend Pseudo-Random Number generator downloaded from the NSL Ariel web site and check the Central Limit Theorem:

1. <font color="red">Add two probability distributions</font> by using the **method of the inversion of the cumulative distribution** to sample from a **generic** <font color="red">exponential distribution</font>, $p(x) = \lambda \exp(-\lambda x)$, $x\in [0;+\infty]$ (see <a href="https://en.wikipedia.org/wiki/Exponential_distribution">this Wikipedia link</a>), and a **generic** <font color="red">Cauchy-Lorentz distribution</font> $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$, $x\in [-\infty;+\infty]$ (see <a href="https://en.wikipedia.org/wiki/Cauchy_distribution">this Wikipedia link</a>).
2. <font color="red">Make 3 pictures</font> with the histograms obtained filling them with $10^4$ realizations of $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ (for $N=1, 2, 10, 100$), being $x_i$ a random variable sampled throwing a *standard* dice (fig.1), an *exponential* dice (fig.2, use $\lambda=1$) and a *Lorentzian* dice (fig.3, use $\mu=0$ and $\Gamma=1$).

Note tha you can try to fit the case $N=100$ with a Gaussian for standard and exponential dices, whereas you should use a Cauchy-Lorentz distribution for the last case.

Il programma implementato per risolvere questo esercizio ha la stessa struttura di quello scritto per l'esercizio 1.1.

Per soddisfare la prima richiesta nel file *random.cpp* sono stati definiti due nuovi metodi: *Exp($\lambda$)* e *Lorentz($\Gamma,\mu$)*. Essi restituiscono un numero casuale proveniente da una distribuzione esponenziale nel primo caso e lorentziana nel secondo. Questi metodi sono stati implemantati utilizzando il metodo della **Funzione cumulativa inversa** che permette di trasformare $x\in[0,1]$ di una distribuzione uniforme in un numero y proveniente da una distribuzione $p_y(y)$ tali che:
$$p_y(y)= dF_y(y)/dy \hspace{1 cm} con \hspace{1 cm} y=F^{-1}_y(x)$$

Per realizzare la seconda richiesta nell'esecuzione del *main()* vengono prodotti tre file. In ciascuno si scrivono $4\cdot10^4$ medie di numeri generati dalla distribuzione desiderata (Gaussiana,Esponenziale,Lorentziana). I primi $10^4$ sono la media ottenuta usando 1 solo numero, i secondi $10^4$ utilizzano 2 numeri, i terzi 10, i quarti 100.
Questi file vengono poi letti dal programma Python qui sotto che ne costruisce un istogramma (trasformato poi in una funzione per poter eseguire un fit) i **grafici** prodotti da questo codice sono raccolti in un unica finestra ciascuna con 4 subplot, uno per ogni numero utilizzato a formare la media (N=1,2,10,100). I **fit** come richiesto sono fatti con una Gaussiana per le distribuzioni Normale e Esponenziale e con una Lorentziana nel terzo caso.

Osserviamo solo che tutte le distribuzioni ottenute per N=10,100 confermano perfettamente la tesi del **teorema del limite centrale** per cui le medie ottenute dalle prime due distribuzioni si dispongono su una distribuzione Gaussiana centrata nel valore atteso della distribuzione originaria con varianza ${\sigma^2}/{\sqrt{N}}$ (sempre più stretta al crescere dei numeri utilizzati per formare una media). Per le distribuzionidi Lorentziane invece si applica la generalizzazione del teorema del limeite centrale che permette di dire che le medie di variabili lorenziane si distrubuiscono lungo una lorentziana a loro volta; in questo caso quindi il fit è stato fatto con una Lorentziana

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
fig=plt.figure(figsize=(16, 5))


def f(x,mu,sigma):
	return 1./np.sqrt(2.*np.pi*sigma**2)*np.exp(-0.5*(x-mu)**2/sigma**2)

N=[1,2,10,100]
for i in range(0,4):
	rows=i*10000
	dati = np.loadtxt("./code/Risultati_Exp.dat", skiprows=rows, max_rows=10000)
	n,bins,paches=plt.hist(dati,100, density=True, facecolor='w')

	x = bins[0:100]
	y = n

	p_opt, p_cov = curve_fit(f, x, y, bounds=([0,0],[2,100]))
	y_fit = f(x,p_opt[0],p_opt[1])

	print ("N =",str(N[i]),": [mu, sigma] = ",p_opt)
	print ("        incertezza = ",np.sqrt(np.diagonal(p_cov)),'\n')

	plt.subplot(1,4,i+1);
	plt.plot(x,y_fit, 'r--') # plotting fitted function
	plt.plot(x,y, 'g')
	plt.title("N="+str(N[i]))
	plt.grid(True)
plt.suptitle("Exponential distribution")
plt.show()
```

```python
fig=plt.figure(figsize=(16, 5))

def f(x,mu,sigma):
	return 1./np.sqrt(2.*np.pi*sigma**2)*np.exp(-0.5*(x-mu)**2/sigma**2)


N=[1,2,10,100]
for i in range(0,4):
	rows=i*10000
	dati = np.loadtxt("./code/Risultati_Gauss.dat", skiprows=rows, max_rows=10000)
	n,bins,paches=plt.hist(dati,100, density=True, facecolor='w')

	x = bins[0:100]
	y = n

	p_opt, p_cov = curve_fit(f, x, y, bounds=([0,0],[2,100]))
	y_fit = f(x,p_opt[0],p_opt[1])

	print ("N =",str(N[i]),": [mu, sigma] = ",p_opt)
	print ("        incertezza = ",np.sqrt(np.diagonal(p_cov)),'\n')

	plt.subplot(1,4,i+1);
	plt.plot(x,y_fit, 'r--') # plotting fitted function
	plt.plot(x,y, 'g')
	plt.title("N="+str(N[i]))
	plt.grid(True)
plt.suptitle("Gaussian distribution")
plt.show()
```

```python
fig=plt.figure(figsize=(16, 5))

def f(x,mu,gamma):
	return gamma/(np.pi*((x-mu)**2.+gamma**2.))

N=[1,2,10,100]
for i in range(0,4):
	rows=i*10000
	dati = np.loadtxt("./code/Risultati_Lor.dat", skiprows=rows, max_rows=10000)
	n,bins,paches=plt.hist(dati,100,range=[-8,8], density=True, facecolor='w')

	x = bins[0:100]
	y = n

	p_opt, p_cov = curve_fit(f, x, y)
	y_fit = f(x,p_opt[0],p_opt[1])

	print ("N =",str(N[i]),": [mu, gamma] = ",p_opt)
	print ("        incertezza = ",np.sqrt(np.diagonal(p_cov)),'\n')

	plt.subplot(1,4,i+1);
	plt.plot(x,y_fit, 'r--') # plotting fitted function
	plt.plot(x,y, 'g')
	plt.title("N="+str(N[i]))
	plt.grid(True)
plt.suptitle("Cauchy-Lorentz distribution")
plt.show()
```
