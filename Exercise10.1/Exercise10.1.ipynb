Il programma implementato per risolvere l'esercizio ha una struttura simile al'esercizio 9.1. E' composto da tre file: *CityGnerator.cpp* che genera la configurazione delle città da raggiungere, *random.cpp* che contiene i metodi per generare numeri casuali e *main.cpp* che contiene il codice per risolvere il problema del TSP come richiesto.

Nel *main.cpp* viene inizializato il programma caricando dal file *main.h* il numero di città *NCity* l'inverso della temperatura iniziale *beta*, il numero di temperature che si esploreranno *NTemp*, il numero di modifiche proposte per ogni temperatura *NStep*. Viene quindi generato un percorso casuale che unisce le città *Path*. La ricerca del percoso migliore prevede che  venga proposta una modifica del percorso a cui si è arrivati con la chiamata della funzione *Mutation*, utilizzando l'algoritmo di Metropolis si decide se rifiutare o accettare la mossa.
La probabilità di transizione utilizzata è:

$$ P = e^{-\beta(Length_{(Proposed Path)}-Length_{(Actual Path)})} $$

Dopo *NStep* con lo stesso *beta*, inizialmente fissato a 1 si aumenta il suo valore in modo da diminuire la probabilità di accettare nuove mosse con la legge $\beta \>\cdot= 1.003$. In questo modo la crescita della temperatura è sufficientemente lenta da permettere al sistema di espolare una vasta zona dello spazio dei parametri durante i primi tentativi.
Il programma salva sul file *Best.dat* per ogni percorso visitato il *beta* con cui è stato ottenuto e la sua lunghezza.

Il set di parametri utilizzati nelle simulazioni sono: $\> NCity=30 \quad  NTemp=2000 \quad NStep1000$

I grafici di seguito rappresentano la lunghezza del percorso in ogni step e il valore di beta in ogni step.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure(figsize=(10, 5))

beta,best = np.loadtxt("code/Best.dat", usecols=(0,1), delimiter='	', unpack='true')

plt.subplot(2, 1, 1)
plt.plot(best, '.-')
plt.xlabel('Numers of population')
plt.ylabel('Length')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(beta, '.-')
plt.xlabel('Numers of population')
plt.ylabel('Beta')
plt.grid(True)

plt.tight_layout()
#plt.savefig("./pictures/BestTripsLength.png")

plt.show()
```
