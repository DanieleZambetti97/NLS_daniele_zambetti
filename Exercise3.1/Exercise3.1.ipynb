Black-Scholes theory assumes that the time evolution of the asset price is stochastic and exhibits geometric Brownian motion (GBM) with constant risk-free interest rate $r$ and volatility $\sigma$.

By using the following parameters: 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

compute at time $t=0$ **via Monte Carlo** the European call-option price, $C[S(0),0]$, and put-option price, $P[S(0),0]$:

1. <span style="color:red"> By sampling **directly** the final asset price $S(T)$</span> for a $GBM(r,\sigma^2)$
2. <span style="color:red"> By sampling the discretized $GBM(r,\sigma^2)$ path of the asset price</span> 
dividing $[0,T]$ in $10^2$ time intervals: $S(0)$, $S(t_1)$, $S(t_2)$, ... up to $S(t_{100}=T=1)$

<span style="color:red">Make four pictures for the estimation of the European call-option prices, $C[S(0),0]$ (direct and discretized), and put-option prices, $P[S(0),0]$ (direct and discretized), with their uncertainties </span> with a large number of asset prices at time $t=T$, say $M\ge 10^4$, as a function of the number of blocks, $N$</font>. <span style="color:blue"> As usual, in computing statistical uncertainties, use data blocking.</span>

In questo esercizio è stato scritto un programma che simula il moto browniano geometrico dei prezzi di un asset, per poter stimare il valore di un obbligazione. Tale valore viene calcolato nel *main()* $M= 10^4$ volte e i dati così ottenuti vengono suddivisi in $N=100$ blocchi ed elaborati con la tecnica del *datablocking*. L'asset che si va a simulare ha prezzo inizale $S(0)=100$ e ne si vuole predirre il prezzo a un tempo $T=1$: l'opzione di cui si vuole stimare il valore ha strike price $K=100$, interesse fisso $r=0.1$ e volabilità $\sigma=0.25$.

I quattro grafici riportati mostrano l'andamento del prezzo di un obbligazione europea con opzione di tipo *Put* i primi due e *Call* i secondi. La differenza tra le due sta nel confontare il massimo tra $(0, K-S(T))$ per le *Put* e tra $(0, S(T)-K)$ per le *Call*, dove $S(T)$ rappresenta il prezzo dell'asset dopo un tempo $T$. Il prezzo finale dell'obbligazione viene quindi stimato come: $$P(S(T),T)= e^{-rT}max(0, K-S(T))$$ $$C(S(T),T)= e^{-rT} max(0, S(T)-K)$$

Per entrambe le opzioni il prezzo finale dell'asset $S(T)$ viene stimato con due metodi differenti: in un caso si calcola direttamnte il prezzo al tempo T, nell'altro si stima l'andamento del prezzo ad intervalli temporali intermedi $S(t_i)$. Nel primo caso si utilizza la funzione *GBM_Direct* che restituisce: $$S(T) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)T+\sigma W(T)\right]$$ dove $W(T)$ è una variabile gaussiana $\mathcal{N}(0,t)$. Nel secondo caso si utilizza per stimare il prezzo dell'asset al passo temporale successivo: $$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
Le due modalità dovrebbero essere equivalenti perché la probabilità congiunta del processo continuo (calcolo diretto) o discreto sono identiche; nei grafici questo è confermato dal fatto che i dati provnienti da entrambi i metodo sono tutti compatibili.
