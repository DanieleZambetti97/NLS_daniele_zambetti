### Exercise 02.1
- Compute the following 1D integral via Monte Carlo 
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$



1. sampling a uniform distribution in $[0,1]$
2. using importance sampling (i.e. sampling a non-uniform probability in $[0,1]$)

<span style="color:red">Show a picture of both your estimations of $I$</span> and their uncertainties with a large number of *throws* $M$ (e.g. $M\ge 10^4$) as a function of the number of blocks, $N$</font>.

Il programma implementato per risolvere questo esercizio è composto da un *main()* e due file *random.cpp* e *datablocking.cpp* in cui sono definite alcune funzioni di supporto.

Nel *main()* dopo aver definito le variabili utilizzate ed inizializzato il generatore di numeri casuali, vengono eseguiti due cicli for. Quello più esterno cicla sul numero di blocchi, quello più interno valuta la funzione integranda in 100 punti casuali uniformemente distribuiti tra [0,1] (intervallo di integrazione) e ne fa la somma. Nel ciclo più esterno viene poi salvata la media dei numeri sommati nel vettore *AV*. Questo viene poi passato alla funzione *DataBlocking* che calcola la miglior stima dell'integrale e il suo errore in funzione del numero di blocchi. I risultati vengono quindi salvati sul file *RisultatiUniforme.dat*.

Nella seconda parte del *main()* si trovano altri due cicli for e un'altra chiamata della funzione *DataBlocking* che salva i risultati ottenuti sul file *RisultatiOttimiza.dat*. Questo secondo conto è identico a quello descritto precedentemente tranne che per la generazione dei punti in cui valutare l'integranda. Ora si è fatto un **importance sampling** per rendere più efficente la generazione di questi. La distribuzione scelta è $p(x) = 2-2x$ perché:
- è una retta decrescente il cui andamento approssima quello dell'integranda meglio di una funzione costante. Entrambe sono decrescenti in [0,1], con valore nullo in x=1.
- l'area sotto la curva in [0,1] risulta 1 in modo che la distribuzione sia normalizzzata.
- la funzione cumulativa di questa distribuzone è facilmente invertibile nell'intervallo di integrazione. Questo ci permette di generare numeri con distribuzione p(x) a partire da una distribuzione uniforme.

Qui sotto è implementato un programma in python che espone i risutltati ottenuti. Osserviamo solamente che utilizzando l'importance sampling si ottengono anche con pochi blocchi stime più accurate che nel caso di una distribuzione uniforme

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x1, f1, error1 = np.loadtxt("./code/RisultatiUnifrome.dat", usecols=(0,1,2), delimiter='	', unpack='true')
plt.errorbar(x1, f1 , yerr=error1, fmt='g.', capsize=1, ecolor= 'green', elinewidth=1, ms=3, label='Distribuzione unifrome')

x2, f2, error2 = np.loadtxt("./code/RisultatiOttimiza.dat", usecols=(0,1,2), delimiter='	', unpack='true')
plt.errorbar(x2, f2 , yerr=error2, fmt='r.', capsize=1, ecolor= 'red', elinewidth=1, ms=3, label='Distribuzione ottimizzata')

plt.xlabel('number of throws')
plt.ylabel('<I>')
plt.title(r"$\int_0^1 \frac{\pi}{2} cos(\frac{\pi}{2} x)\,dx$")
#'Stima integrale I=pi/2 * cos(pi/2*x)'
plt.legend()
plt.grid(True)

plt.show()
```
