### Exercise 04.4

By means of your upgraded MD code, equilibrate and <span style="color:red">perform MD simulations via a Lennard-Jones model</span> of Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) and Krypton ($\sigma = 0.364$ nm, $\epsilon/k_B = 164$ K, $m=83.798$ amu) in the following conditions:
1. solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

and show in pictures the obtained average values and uncertainties for the total (internal) energy per particle, $E/N$, the potential energy per particle, $U/N$, the kinetic energy per particle, $K/N$, the pressure $P$ and the temperature, $T$ ... in SI units!

Il programma scritto per risolvere questo esercizio è costituito da uno script in python *equilibration.py* e dal codice utilizzato nell'esecizio 4.3. I grafici ottenuti dalle simulazioni richieste sono stati salvati nella cartella *picture* e vengono qui riportati.

Ogni stato termodinamico è stato simulato con un iniziale fase di equilibrazione del sistema eseguito proprio dallo script di python. Questo inizia una nuova simulazione partendo dalla configurazione *fcc* fornita e fa compiere al sistema $10^3$ step temporali misurando le osservabili richieste ogni $10$ mosse. Le misure istantanee delle osservabili sono poi raccolte in $10$ blocchi da $10$ misure l'uno e con essi viene svolto un datablocking che permette di stabilire il valore medio di ciscuna osservabile. Lo script *equilibration.py* procede a ricominciare la simulazione mediante l'opzione di *restart* passata da riga di comando fino a quando la stima della temperatura media ottenuta non diventa compatibile con la temperatura deisderata $\pm 0.05$.

Una volta terminata la fase di equilibrazione è stato modificato il file *input.dat* per eseguire una simulazione più lunga di $10^4$ step temporali, nelle quali si misurano i valori delle osservabili di interesse ogni $10$ mosse, le $10^3$ misure istantanee vengono poi raggruppate in $10^2$ blocchi. Mediante il comando *./MolDyn_NVE.x 1*  si lancia questa nuova simulazione.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(13, 8))
data = np.loadtxt("./code/risultati/output_temp.dat")
Maxrows = 10**4

Etot = np.loadtxt("./code/risultati/output_etot.dat", max_rows= Maxrows)
Epot = np.loadtxt("./code/risultati/output_epot.dat", max_rows= Maxrows)
Ekin = np.loadtxt("./code/risultati/output_ekin.dat", max_rows= Maxrows)
Temp = np.loadtxt("./code/risultati/output_temp.dat", max_rows= Maxrows)
Pres = np.loadtxt("./code/risultati/output_pres.dat", max_rows= Maxrows)

plt.subplot(2,1,1)
plt.plot(Temp, '.-', markersize=2, label="Temperature" )
plt.title("Gas Phase")
plt.xlabel("Time Step")
plt.ylabel("Temperature")
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(Etot, '.-', markersize=2, label="Total Energy" )
plt.title("Gas Phase")
plt.xlabel("Time Step")
plt.ylabel("Total Energy")
plt.grid()

plt.tight_layout()
plt.savefig("./pictures/gas_equilibration.png")
plt.show()
```

```python
fig=plt.figure(figsize=(16, 10))
sigma=0.364
e_su_kb=164
er=1
plt.subplot(3,2,1)
x, f, error = np.loadtxt("./code/risultati/ave_temp.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f*e_su_kb,yerr=error*e_su_kb, errorevery=er, elinewidth=1, capsize=1, ms=3)
plt.xlabel('Numers of block')
plt.ylabel("T [K]")
plt.title("Data Blocking: Temperature")
plt.grid(True)

plt.subplot(3,2,2)
x, f, error = np.loadtxt("./code/risultati/ave_etot.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f*e_su_kb*1.38**-23,yerr=error*e_su_kb*1.38**-23, errorevery=er, elinewidth=1, capsize=1, ms=3)
plt.xlabel('Numers of block')
plt.ylabel("Etot/N [J]")
plt.title("Data Blocking: Total Energy")
plt.grid(True)

plt.subplot(3,2,3)
x, f, error = np.loadtxt("./code/risultati/ave_epot.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f*e_su_kb*1.38**-23,yerr=error*e_su_kb*1.38**-23, errorevery=er, elinewidth=1, capsize=1, ms=3)
plt.xlabel('Numers of block')
plt.ylabel("Epot/N [J]")
plt.title("Data Blocking: Potential Energy")
plt.grid(True)

plt.subplot(3,2,4)
x, f, error = np.loadtxt("./code/risultati/ave_ekin.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f*e_su_kb*1.38**-23,yerr=error*e_su_kb*1.38**-23, errorevery=er, elinewidth=1, capsize=1, ms=3)
plt.xlabel('Numers of block')
plt.ylabel("Ekin/N [J]")
plt.title("Data Blocking: Kinetic Energy")
plt.grid(True)

plt.subplot(3,2,5)
x, f, error = np.loadtxt("./code/risultati/ave_pres.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f*e_su_kb*1.38**-23/(sigma*10**-9)**3,yerr=error*e_su_kb*1.38**-23/(sigma*10**-9)**3, errorevery=er, elinewidth=1, capsize=1, ms=3)
plt.xlabel('Numers of block')
plt.ylabel("Pressure [Pa]")
plt.title("Data Blocking: Pressure")
plt.grid(True)

plt.tight_layout()
plt.savefig("./pictures/gas_datablocking_Krypton.png")
plt.show()
```
