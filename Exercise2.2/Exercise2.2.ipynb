- <span style="color:red">3D Random Walks (RW) on a cubic lattice and in the continuum:</span> Repeat many times (e.g. say $10^4$) the simulation of a random walk in 3D always starting at the origin

1. on a cubic lattice with lattice constant $a=1$; at each discrete time the walker makes a forward or backward step of length equal to $a$ in one of the 3 principal directions of the lattice: $x$, $y$ or $z$
2. in the continuum; at each discrete time the walker makes a step of length equal to $a(=1)$ along a **random direction** obtained by sampling **uniformly** the solid angle: $\theta \in [0,\pi]$ and $\phi \in [0,2\pi]$

<span style="color:red">Show a picture of $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ for both RWs</span>, with their statistical uncertainties, as function of the step $i\in [0,10^2]$.

Note that you could try to fit both results with a function like $f(N)=k\sqrt{N}$. Do your results indicate a diffusive behavior?

Per risolvere l'esercizio si è scritto un programma contenuto nel file *main.cpp*.  All'interno di questo viene inizializzato il generatore di numeri casuali, si definisce la funzione *NewPos* e si implementa il *main()*.

Quest'ultimo è pensato per riempire un vettore lungo $nStep = 100$ con $nWalk=10^4$ realizzazioni di un Random Walk. A ogni step di ogni cammino viene sommata la distanza  a cui si è arrivati rispetto all'origine e il suo quadrato.
Terminati tutti gli spostamenti si calcola per ogni step $i\in [0,10^2]$ la media su tutti i cammini, l'errore statistico di tale media e si salvano i risultati su un file.

Tutto questo processo è ripetuto due volte: la prima volta per calcolare la posizione successiva in un Random Walk si utilizza un passo discreto, la seconda uno continuo.

Alcune osservazioni:
- Per generare una direzione casuale in tre dimensioni in manienera uniforme si generano punti casuali in un cubo di lato $[-1,1]$ scartando i punti che non sono contenuti nella sfera inscritta dentro il cubo.
- L'esercizio richiedeva di graficare $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$. Il nostro programma non misura direttamente questa quantità quindi per calcolarene l'errore statistico bisogna utilizzare la propoagazione degli errori. Tale operazione è svolta nel main quando si scrive su file la distanza media ottenuta e il suo errore. La formula utilizzata è:
$$ error_{\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}}= \frac{error_{\langle |\vec{r}_N|^2 \rangle_{RW}}}{0.5*{\langle |\vec{r}_N|^2 \rangle_{RW}}}$$

Qui sotto riportiamo il grafico della miglior stima della distanza dall'origine ottenuta con il datablocking in funzione dei passi fatti nei due casi di passo discreto e continuo. I dati sono fittati con una funzione $f(N)=k\sqrt{N}$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

def f(x,k):
	return k*np.sqrt(x)

x1, f1, error = np.loadtxt("./code/Risultati0.dat", usecols=(0,1,2), delimiter='	', unpack='true')
p_opt, p_cov = curve_fit(f, x1, f1)
y_fit = f(x1,p_opt[0])

plt.errorbar(x1,f1, yerr=error, label="Dati")
plt.plot(x1,y_fit, label="Fit", color='r') # plotting fitted function

plt.xlabel('passi fatti')
plt.ylabel('<dist>')
plt.title("Discrete step")
plt.grid(True)
plt.legend()
print("optimized parameters k=")
print(p_opt)
print("parameters uncertainty =")
print(np.sqrt(np.diagonal(p_cov)))
plt.show()
```
