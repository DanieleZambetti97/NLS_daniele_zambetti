### Exercise 08.1

Write a C++ Variational Monte Carlo code for a single quantum particle in 1D which exploits the Metropolis algorithm to sample a trial wave function $|\Psi_T(x)|^2$ using an uniform transition probability $T(x_{new}|x_{old})$ (Suggestion: as a starting point, you can use the code you have already developed for the Hydrogen atom or the Monte Carlo NVT code). Using data blocking, the code should be able to compute the expectation value for the Hamiltonian

$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$

which can be used to find the parameters that minimize this quantity.

Il codice implementato per risolvere questo esercizio utilizza l'algoritmo di metropolis per campionare l'energia potenziale di uno stato descritto da una funzione d'onda del tipo
$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$
Il programma deve essere eseguito lanciando da terminale il comando *./main.exe mu sigma*. Vengono calcolati $10^6$ valori di energia potenziale del sistema utilizzando come stato la $\Psi^{\sigma,\mu}(x)$ con i valori dei parametri passati da riga di comando. Questi sono raccolti poi in $100$ blocchi e analizzati con la tecnica del datablocking. Il risultato di questa analisi (valor medio di energia potenziale e suo errore) sono poi stampati nel file *Energie.dat*.

Di seguito riportiamo l'andamento della stima dell'energia potenziale in funzione del numero di blocchi utilizzati per i valori dei paramentri $\mu=1$  $\sigma=1$

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,5))

i, error, data = np.loadtxt('./code/Risultati.dat', usecols=(0,1,2),delimiter='	', unpack='true')

plt.errorbar(i,data, yerr=error)
plt.title("Data blocking: Potential energy")
plt.xlabel("Number of block")
plt.ylabel("Epot")
plt.grid()
plt.show()
```
