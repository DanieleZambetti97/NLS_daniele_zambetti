Questo codice mostra all'interno di una signola simulazione l'andamento delle osservabili di interesse al crescere del numero di blocchi utilizzati per stimarle. Tali osservabili dovrebbero convergere verso il valore teorico predetto per la temperatura a cui si Ã¨ eseguita la simulazione (si veda il jupyter notebook Exercise6.1)

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#fig=plt.figure(figsize=(15, 10))

x, f, error = np.loadtxt("./code/risultati/DataBlocking_Ene.dat", usecols=(0,1,2), delimiter='	', unpack='true')
plt.errorbar(x,f,yerr=error, errorevery=1, fmt='c.-', elinewidth=1, capsize=1, ms=3, label="Gibbs")

plt.xlabel('Number of block')
plt.ylabel('$U/N$')
plt.title('Internal Energy')
plt.grid(True)

plt.show()
```
